This notebook requires the following imports:

```python
from rnet.toolkits.coords import *

from itertools import product
import matplotlib.pyplot as plt
import numpy as np
```

# Points within a circle

Given the $x$- and $y$-coordinates of a grid of points, `indices_in_circle()` returns the indices of the points that are located with a circle centered at $(x,y)$ with radius $r$.

```python
# Example parameters
xdata = np.arange(-3,4,0.5)
ydata = np.arange(5,0,-1)
xc, yc, r = 0.25, 3, 1.4
# Compute the indices
i,j,d = indices_in_circle(xdata,ydata,xc,yc,r)
# Plot the results
coords = np.array(list(product(xdata,ydata)))
plt.scatter(*coords.T,s=10,color='k')
plt.scatter(xdata[i],ydata[j],s=10,color='tab:red')
plt.scatter(xc,yc,s=30,color='tab:blue')
t = np.linspace(0,2*np.pi,720)
plt.plot(xc+r*np.cos(t),yc+r*np.sin(t),color='tab:blue',linestyle='--')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.tight_layout()
plt.tick_params(direction='in',top=True,right=True)
plt.show()
```
